```python
import matplotlib.pyplot as plt

exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

plt.figure(figsize=(10, 6))
plt.hist(exp_data, bins=100, density=True, histtype='step', label='Exponential', color='green')
plt.hist(uniform_data, bins=100, density=True, histtype='step', label='Uniform', color='blue')
plt.hist(normal_data, bins=100, density=True, histtype='step', label='Normal', color='red')
plt.title('Random Distributions')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title("3D Plot of Z = X² + Y²")
plt.show()
```

```python
# Task 2 – Au Nano Particle Dataset Analysis
# ---------------------------------------------
# This notebook loads Au_nanoParticle_dataset.csv and performs analysis using pandas, matplotlib, and seaborn.

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. LOAD THE DATASET
# -------------------------------------------------
# Upload Au_nanoParticle_dataset.csv in Colab using the left-hand panel OR mount Google Drive.
df = pd.read_csv("Au_nanoparticle_dataset.csv")

print("Dataset Loaded Successfully!")
print("Columns available:", df.columns.tolist())
print("Shape of dataset:", df.shape)

# 2. FILTER THE REQUIRED COLUMNS
# -------------------------------------------------
new_df = df[["N_total", "N_bulk", "N_surface", "R_avg"]]
print("\nFiltered DataFrame Columns:", new_df.columns.tolist())

# 3. DISPLAY THE FIRST 20 SAMPLES
# -------------------------------------------------
print("\nFirst 20 Samples:")
print(new_df.head(20))

# 4. CALCULATE MEAN, STANDARD DEVIATION & QUARTILES
# -------------------------------------------------
stats = new_df.describe()  # gives count, mean, std, min, 25%, 50%, 75%, max
print("\nSummary Statistics for Selected Features:")
print(stats)

# 5. PLOT HISTOGRAMS FOR ALL 4 FEATURES IN 1x4 LAYOUT
# -------------------------------------------------
plt.figure(figsize=(20,5))

for i, column in enumerate(new_df.columns):
    plt.subplot(1, 4, i+1)
    plt.hist(new_df[column], bins=20, color="skyblue", edgecolor="black")
    plt.title(f"Histogram of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

# 6. SEABORN PAIRPLOT (Scatter plots + histograms)
# -------------------------------------------------
sns.pairplot(new_df)
plt.suptitle("Pairplot of Au Nano Particle Features", y=1.02)
plt.show()

# 7. CUSTOMIZED PAIRGRID
# -------------------------------------------------
g = sns.PairGrid(new_df)

# Upper half → Bivariate histogram
g.map_upper(sns.histplot)

# Diagonal → Histogram + KDE (Kernel Density Estimation)
g.map_diag(sns.histplot, kde=True)

# Lower half → Bivariate KDE plot (filled)
g.map_lower(sns.kdeplot, fill=True)

plt.suptitle("Customized PairGrid with KDE & Histograms", y=1.02)
plt.show()
```
